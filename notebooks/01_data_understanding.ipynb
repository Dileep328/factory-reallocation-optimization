## Initial Business Understanding

The dataset contains order-level transactional data
for Nassau Candy Distributor.

Key business dimensions:

1. Product
2. Region
3. Ship Mode
4. Sales
5. Gross Profit
6. Cost

The objective is to build a system that can recommend
optimal factory-product assignments to improve shipping
efficiency and profitability.

### Key Findings

- Dataset contains 10,194 records.
- Dataset contains 15 products.
- Dataset contains 4 regions.
- Dataset contains 4 ship modes.
- No missing values detected.
- No major data quality issues observed.
- Product-factory mapping will need to be added externally.
