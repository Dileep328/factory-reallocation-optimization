```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv(
    "../data/processed/cleaned_data.csv"
)
```

```python
df_fe = df.copy()
```

# Feature 1

```python
df_fe['Order Date'] = pd.to_datetime(
    df_fe['Order Date']
)
```

```python
df_fe['Order_Month'] = (
    df_fe['Order Date']
    .dt.month
)
```

# Feature 2

```python
df_fe['Order_Quarter'] = (
    df_fe['Order Date']
    .dt.quarter
)
```

# Feature 3

```python
df_fe['Margin_Category'] = pd.cut(
    df_fe['Profit_Margin'],
    bins=[-1,0.1,0.3,1],
    labels=[
        'Low',
        'Medium',
        'High'
    ]
)
```

# Feature 4

Sales Per Unit  

```python
df_fe['Sales_Per_Unit'] = (
    df_fe['Sales']
    /
    df_fe['Units']
)
```

# Feature 5 

```python
Profit per unit 
```

```python
df_fe['Profit_Per_Unit'] = (
    df_fe['Gross Profit']
    /
    df_fe['Units']
)
```

# Feature 6

```python
factory_load = (
    df_fe.groupby('Factory')
    .size()
    .reset_index(
        name='Factory_Load'
    )
)
```

```python
df_fe = df_fe.merge(
    factory_load,
    on='Factory'
)
```

# Feature 7

```python
Product demand 
```

```python
product_demand = (
    df_fe.groupby(
        'Product Name'
    )
    .size()
    .reset_index(
        name='Product_Demand'
    )
)
```

```python
df_fe = df_fe.merge(
    product_demand,
    on='Product Name'
)
```

# Feature 8

```python
regional_demand = (
    df_fe.groupby('Region')
    .size()
    .reset_index(
        name='Regional_Demand'
    )
)
```

```python
df_fe = df_fe.merge(
    regional_demand,
    on='Region'
)
```

# Feature 9

```python
df_fe['Route'] = (
    df_fe['Factory']
    +
    "_"
    +
    df_fe['Region']
)
```

# Feature 10

```python
route_volume = (
    df_fe.groupby('Route')
    .size()
    .reset_index(
        name='Route_Volume'
    )
)
```

```python
df_fe = df_fe.merge(
    route_volume,
    on='Route'
)
```

# Feature 11

```python
df_fe['Product_Frequency'] = (
    df_fe.groupby(
        'Product Name'
    )['Product Name']
    .transform('count')
)
```

# Feature 12

```python
factory_eff = (
    df_fe.groupby('Factory')
    ['Lead_Time']
    .mean()
)
```

```python
factory_eff = (
    factory_eff
    .reset_index()
)
```

```python
factory_eff.columns = [
    'Factory',
    'Factory_Efficiency'
]
```

```python
df_fe = df_fe.merge(
    factory_eff,
    on='Factory'
)
```

# Encoding 

```python
le_factory = LabelEncoder()

df_fe['Factory_Encoded'] = (
    le_factory.fit_transform(
        df_fe['Factory']
    )
)
```

```python
le_region = LabelEncoder()

df_fe['Region_Encoded'] = (
    le_region.fit_transform(
        df_fe['Region']
    )
)
```

```python
le_ship = LabelEncoder()

df_fe['ShipMode_Encoded'] = (
    le_ship.fit_transform(
        df_fe['Ship Mode']
    )
)
```

```python
le_product = LabelEncoder()

df_fe['Product_Encoded'] = (
    le_product.fit_transform(
        df_fe['Product Name']
    )
)
```
